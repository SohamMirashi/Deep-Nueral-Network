```python
import numpy as np
```

```python
# parameters
m = 2.5      # slope
c = 1.5      # intercept
noise_level = 2.3462
```

```python
# create 100 x-values (e.g. spanning 0 to 10)
x = np.linspace(0, 50, 100)
```

```python
# create noise term
noise = np.random.normal(loc=0, scale=noise_level, size=100)
```

```python
# compute y values
# y = m * x + c + noise
# y = (m*x + c + noise).astype(int)
y = (m*x + c + noise)
```

```python
print(x, "\n\n")
print(y)
```

```python
import matplotlib.pyplot as plt

# plt.scatter(x, y)
# plt.xlabel("x values")
# plt.ylabel("y values")
# plt.title("Scatter Plot of y = mx + c + noise")
# plt.show()

alpha = 0.0001
w = 0.0
b = 0.0
epochs = 3000

for _ in range(epochs):
    for xi, yi in zip(x, y):
        y_pred = w * xi + b
        error = yi - y_pred
        w += alpha * error * xi
        b += alpha * error

print("Learned w:", w)
print("Learned b:", b)

# Plot results
plt.scatter(x, y, label="Data")
y_pred_line = [w*xi + b for xi in x]
plt.plot(x, y_pred_line, color='red', label="Regression Line")
plt.legend()
plt.show()
```
