```python
import numpy as np
```

```python
# parameters
m = 2.5      # slope
c = 1.5      # intercept
noise_level = 2.3462
```

```python
# create 100 x-values (e.g. spanning 0 to 10)
x = np.linspace(0, 50, 100)
```

```python
# create noise term
noise = np.random.normal(loc=0, scale=noise_level, size=100)
```

```python
# compute y values
# y = m * x + c + noise
y = (m*x + c + noise).astype(int)
```

```python
print(x, "\n\n")
print(y)
```

```python
import matplotlib.pyplot as plt

plt.scatter(x, y)
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("Scatter Plot of y = mx + c + noise")
plt.show()
```

```python
from faker import Faker

fake = Faker()

# Parameters
m = 2
c = 1
noise_level = 3

# Generate 100 fake integer x-values
x = [fake.random_int(min=0, max=150) for _ in range(100)]

# Generate matching noise values
noise = [fake.random_int(min=-noise_level, max=noise_level) for _ in range(100)]

# Compute y-values using the equation y = mx + c + noise
y = [m * xi + c + ni for xi, ni in zip(x, noise)]

print("x values =", x)
print("y values =", y)
```

```python
import matplotlib.pyplot as plt

plt.scatter(x, y)
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("Scatter Plot of y = mx + c + noise")
plt.show()
```
